```python
import numpy as np
import pandas as pd
import json
import os
import urllib.request
from tqdm import tqdm_notebook
```

```python
annotation_path = os.path.join('Data/data_annotation.csv')
train_folder = os.path.join('train')
```

```python
annotation = pd.read_csv(annotation_path)
annotation.head()
```

```python
annotation = annotation.rename(columns={'annotation/0/label/0': 'class', 'annotation/0/points/0/x': 'xmin',
                                       'annotation/0/points/0/y': 'ymin', 'annotation/0/points/1/x': 'xmax',
                                       'annotation/0/points/1/y': 'ymax', 'annotation/0/imageWidth': 'width',
                                       'annotation/0/imageHeight': 'height'})
annotation.head()
```

```python
used_cols = ['content', 'class', 'xmin', 'ymin', 'xmax', 'ymax', 'width', 'height']
non_used_cols = set(annotation.columns) - set(used_cols)

annotation = annotation.drop(non_used_cols, axis=1)
annotation.head()
```

```python
annotation.content[0].split('/')[-1]
```

```python
for img_url_ind in tqdm_notebook(range(len(annotation.content)), desc='NUMBER OF DOWNLOADED IMAGES'):
    img_url = annotation.content[img_url_ind]
    
    img_name = img_url.split('/')[-1]
    
    urllib.request.urlretrieve(img_url, os.path.join(train_folder, img_name))
```

# PREPARE ANNOTATION

```python
res_names = []
for name in annotation.content:

    name = name.split('/')[-1]
    res_names.append(name)
    
annotation['filename'] = res_names
```

```python
annotation = annotation.drop('content', axis=1)
annotation.head()
```

```python
annotation['xmin'] = annotation['xmin'] * annotation['width']
annotation['xmax'] = annotation['xmax'] * annotation['width']

annotation['ymin'] = annotation['ymin'] * annotation['height']
annotation['ymax'] = annotation['ymax'] * annotation['height']
```

```python
annotation
```

```python
annotation.to_csv('annotation.csv', index=False, index_label=False)
```
